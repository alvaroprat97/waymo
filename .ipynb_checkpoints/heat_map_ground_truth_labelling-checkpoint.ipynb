```python
def get_heat(centre_x, centre_y, height, depth, obect_type, heat_maps, img):
    top_left = (max(0, int(centre_x - 0.5*depth)), 
                max(0, int(centre_y - 0.5*height)))
    bottom_right = (min(heat_maps.shape[2]-1, math.ceil(centre_x + 0.5*depth + 1)), 
                    min(heat_maps.shape[1]-1, math.ceil(centre_y + 0.5*height + 1)))

    x,y = np.meshgrid(np.arange(top_left[0], bottom_right[0]),
                      np.arange(top_left[1], bottom_right[1]))
    pos = np.dstack((x,y))
    mean = np.array([centre_x, centre_y])
    cov  = np.array([[depth**2/4,0],[0,height**2/4]])

    rv = multivariate_normal(mean,cov)
    z = rv.pdf(pos)

    heat_maps[object_type, top_left[1]:bottom_right[1], top_left[0]:bottom_right[0]] = torch.tensor(z)

    plt.imshow(heat_maps[object_type])
    plt.axis('off')
    
    ##### To compare ######
    plt.figure()
    labeled_box = blob['labels'].labels[3]
    ax = plt.subplot()
    ax.add_patch(patches.Rectangle(
    xy=(labeled_box.box.center_x - 0.5 * labeled_box.box.length,
        labeled_box.box.center_y - 0.5 * labeled_box.box.width),
    width=labeled_box.box.length,
    height=labeled_box.box.width,
    linewidth=1,
    edgecolor='red',
    facecolor='none'))
    plt.imshow(imag)
    plt.axis('off')
    return 
```

```python
class image_heat_maps():
    
    def __init__(self, dataset, blob_idx, frame_idx, camera, heat_type):
        
        # Information from image
        self.data_path = os.getcwd() + '/data/' + dataset
        self.blob_idx = blob_idx
        self.frame_idx = frame_idx
        self.camera = camera
        self.path = f'{self.data_path}/{self.camera}/blob_{self.blob_idx}_frame_{self.frame_idx}.pickle'
        self.blob = self._load_blob(self.path)
        
        # Heat maps
        self.img = tf.image.decode_jpeg(self.blob['image'])
        self.img_np = torch.tensor(self.img.numpy())
        self.x_pixels, self.y_pixels = self.img_np.shape[0], self.img_np.shape[1]
        self.heat_maps = torch.zeros((4, self.x_pixels, self.y_pixels))
        self.heat = heat_type
        
    
    def _load_blob(self, path):
        with open(path, 'rb') as pickle_file:
            return pickle.load(pickle_file)
    
    
    def _get_heat(self):
        top_left = (max(0, int(self.centre_x - 0.5*self.depth)), 
                    max(0, int(self.centre_y - 0.5*self.height)))
        bottom_right = (min(self.heat_maps.shape[2]-1, math.ceil(self.centre_x + 0.5*self.depth + 1)), 
                        min(self.heat_maps.shape[1]-1, math.ceil(self.centre_y + 0.5*self.height + 1)))
        
        if self.heat == 'gaussian' or self.heat == 'laplacian':
            x,y = np.meshgrid(np.arange(top_left[0], bottom_right[0]),
                              np.arange(top_left[1], bottom_right[1]))
            pos = np.dstack((x,y))
            mean = np.array([self.centre_x, self.centre_y])
#             cov  = np.array([[self.depth**2/4,0],[0,self.height**2/4]])
            cov = np.array([[self.depth, 0], [0, self.height]])

            rv = multivariate_normal(mean,cov)
            z = rv.pdf(pos)
            
            if self.heat == 'laplacian':
                z = scipy.ndimage.filters.laplace(z)
                z_max = z.max()
                z_min = z.min()
                z = -(z-z_min)/(z_max-z_min) + 1
                z[0,:], z[-1,:], z[:,0], z[:,-1] = 1, 1, 1, 1
#                 z[z<10**(-5)] = 0

        elif self.heat == 'ones':
            z = np.ones((bottom_right[1] - top_left[1], bottom_right[0] - top_left[0]))

        self.heat_maps[self.cls, top_left[1]:bottom_right[1], top_left[0]:bottom_right[0]] = torch.tensor(z)
        return
    
    
    def _get_parameters(self, blob_labels):
        # Labels information
        self.centre_x = blob_labels.box.center_x
        self.centre_y = blob_labels.box.center_y
        self.depth = blob_labels.box.length
        self.height = blob_labels.box.width
        self.cls = blob_labels.type - 1
        return

    def construct_heat_maps(self):
        for blob_labels in self.blob['labels'].labels:
            self._get_parameters(blob_labels)
            self._get_heat()
    
    def return_maps(self):
        return self.heat_maps
    
    def plot_maps(self):
        fig, axs = plt.subplots(2, 2)
        fig.tight_layout(pad=0.5)
        fig.set_size_inches((20,10))
        labels = ['vehicle', 'pedestrian', 'sign', 'background']
        for idx, ax in enumerate(axs.flat):
            ax.imshow(self.heat_maps[idx, :, :])
            ax.set_title(labels[idx])
            
    def plot_image(self):
        plt.imshow(self.img_np)
    
    def plot_image_labels(self):
        ax = plt.subplot()
        for label in self.blob['labels'].labels:
            # Draw the object bounding box.
            ax.add_patch(patches.Rectangle(
            xy=(label.box.center_x - 0.5 * label.box.length,
                label.box.center_y - 0.5 * label.box.width),
            width=label.box.length,
            height=label.box.width,
            linewidth=1,
            edgecolor='red',
            facecolor='none'))
        plt.imshow(torch.tensor(self.img_np))
        plt.axis('off')
```
